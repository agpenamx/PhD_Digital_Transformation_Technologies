## Modelo terico de regresin lineal mltiple

La regresin lineal mltiple permite modelar la relacin entre una variable dependiente y mltiples variables independientes.

La forma general del modelo es:

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_p X_p + \varepsilon
$$

Donde:

$$
\begin{aligned}
Y & = \text{Variable dependiente (respuesta)} \\\\
X_1, X_2, \dots, X_p & = \text{Variables independientes (predictores)} \\\\
\beta_0 & = \text{Intercepto o trmino constante} \\\\
\beta_i & = \text{Coeficiente asociado a } X_i \\\\
\varepsilon & = \text{Error aleatorio del modelo}
\end{aligned}
$$

Para este caso de estudio especfico:

$$
Y = \text{Ventas}, \quad X_1 = \text{TV}, \quad X_2 = \text{Radio}, \quad X_3 = \text{Peridico}
$$

El objetivo del modelo es estimar los parmetros \( \beta_0, \beta_1, \dots, \beta_p \) que mejor predicen \( Y \), **minimizando la suma de los errores cuadrticos** entre los valores observados y los ajustados.

## Forma matricial del modelo de regresin lineal mltiple

El modelo de regresin lineal mltiple tambin puede representarse de forma compacta utilizando lgebra lineal:

$$
\mathbf{Y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\varepsilon}
$$

Donde:

$$
\begin{aligned}
\mathbf{Y} &= 
\begin{bmatrix}
Y_1 \\
Y_2 \\
\vdots \\
Y_n
\end{bmatrix}
\quad \text{(vector de variables dependientes)} \\\\

\mathbf{X} &=
\begin{bmatrix}
1 & X_{11} & X_{12} & \cdots & X_{1p} \\
1 & X_{21} & X_{22} & \cdots & X_{2p} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & X_{n1} & X_{n2} & \cdots & X_{np}
\end{bmatrix}
\quad \text{(matriz de diseo, incluye columna de 1's para el intercepto)} \\\\

\boldsymbol{\beta} &=
\begin{bmatrix}
\beta_0 \\
\beta_1 \\
\vdots \\
\beta_p
\end{bmatrix}
\quad \text{(vector de coeficientes)} \\\\

\boldsymbol{\varepsilon} &=
\begin{bmatrix}
\varepsilon_1 \\
\varepsilon_2 \\
\vdots \\
\varepsilon_n
\end{bmatrix}
\quad \text{(vector de errores aleatorios)}
\end{aligned}
$$

---

### Estimacin de coeficientes por mnimos cuadrados ordinarios (OLS)

Los coeficientes que minimizan la suma de errores cuadrticos se obtienen con la frmula:

$$
\hat{\boldsymbol{\beta}} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{Y}
$$

Esta expresin utiliza la transpuesta de la matriz \( \mathbf{X} \) y su inversa, y es vlida cuando \( \mathbf{X}^\top \mathbf{X} \) es invertible.

```python
X = df[['TV', 'Radio', 'Periodico']]
X = sm.add_constant(X)
y = df['Ventas']
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Residuos y valores ajustados
residuals = model.resid
fitted = model.fittedvalues

# 1. Residuos vs Ajustados con anotaciones
plt.figure(figsize=(10, 6))
plt.scatter(fitted, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuos vs Ajustados")

# Anotacin de patrn en forma de U
plt.text(14, -5.5, "Patrn de 'U'\n posible heterocedasticidad", fontsize=10, color='darkred')
plt.axvspan(10, 20, color='orange', alpha=0.2)

# Anotar un outlier visible
outlier_idx = np.argmax(np.abs(residuals))
plt.annotate(
    f'Outlier {outlier_idx}',
    xy=(fitted[outlier_idx], residuals[outlier_idx]),
    xytext=(fitted[outlier_idx] + 1.5, residuals[outlier_idx] - 2),
    arrowprops=dict(facecolor='black', arrowstyle='->'),
    fontsize=9
)

plt.xlabel("Valores Ajustados")
plt.ylabel("Residuos")
plt.grid(True)
plt.show()

# 2. Q-Q Plot con anotaciones
plt.figure(figsize=(8, 6))
sm.qqplot(residuals, line='45', fit=True)
plt.title("Q-Q plot")

# Texto explicativo sobre desviacin
plt.text(-6, -2.5, "Desviacin de normalidad\ncolas asimtricas", fontsize=10, color='blue')
plt.annotate(
    "Outliers / colas largas",
    xy=(-2, -5),
    xytext=(-5, -6.5),
    arrowprops=dict(facecolor='blue', arrowstyle='->'),
    fontsize=9
)

plt.grid(True)
plt.show()
```

##  Interpretacin del modelo inicial
- **TV** y **Radio** son estadsticamente significativos (p < 0.05)
- **Peridico** no lo es (p = 0.14)

 Se recomienda eliminar Peridico para simplificar el modelo y evitar ruido.
_Este proceso forma parte de feature selection._

## Modelo optimizado (sin Peridico)

```python
X_opt = df[['TV', 'Radio']]
X_opt = sm.add_constant(X_opt)
y_opt = df['Ventas']
model_opt = sm.OLS(y_opt, X_opt).fit()
print(model_opt.summary())
```

```python
resid_opt = model_opt.resid
fitted_opt = model_opt.fittedvalues
plt.scatter(fitted_opt, resid_opt)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuos modelo optimizado')
plt.show()
sm.qqplot(resid_opt, line='45')
plt.title("Q-Q plot modelo optimizado")
plt.show()
```

## Conclusin Final del Modelo Optimizado

Tras evaluar un modelo de regresin lineal mltiple con tres predictores (TV, Radio y Peridico), se encontr que **Peridico no aporta significancia estadstica al modelo** (p = 0.860). Se procedi entonces a optimizar el modelo, conservando solo **TV** y **Radio** como predictores.

---

### Ecuacin del modelo final

$$
\hat{Ventas} = 2.9211 + 0.0458 \cdot TV + 0.1880 \cdot Radio
$$

- Estadstico F: F = 859.6, p < 0.00001  Modelo globalmente significativo.

- Skew: -1.323 | Kurtosis: 6.292  Asimetra negativa y colas pesadas.
