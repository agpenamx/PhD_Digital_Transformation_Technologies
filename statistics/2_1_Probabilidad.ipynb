# Ejercicio 1.19  Probabilidad de orden creciente sin reemplazo

Supn que se eligen $$k$$ nmeros del conjunto $$\{1, 2, 3, \ldots, n\}$$, donde $$k < n$$, sin reemplazo.  
Todos los resultados son igualmente probables.  
Cul es la **probabilidad** de que los nmeros seleccionados estn en **orden creciente**?

## Fundamento terico

Al elegir $$k$$ elementos distintos del conjunto $$\{1, 2, ..., n\}$$ sin reemplazo, el nmero total de posibles subconjuntos es:

$$
\binom{n}{k}
$$

Cada subconjunto tiene $$k!$$ permutaciones posibles (formas distintas de ordenar los mismos $$k$$ nmeros).

De esas permutaciones, **slo una** est en orden estrictamente creciente.

Por lo tanto, la probabilidad de que los $$k$$ nmeros seleccionados estn en orden creciente es:

$$
P = \frac{1}{k!}
$$

## Resolucin paso a paso

La probabilidad de que un subconjunto de $$k$$ nmeros se ordene en forma creciente entre todas sus permutaciones es:

$$
P(\text{orden creciente}) = \frac{1}{k!}
$$

Por ejemplo:

- Si $$k = 3$$:

$$
P = \frac{1}{3!} = \frac{1}{6} \approx 0.1667
$$

- Si $$k = 4$$:

$$
P = \frac{1}{4!} = \frac{1}{24} \approx 0.0417
$$

```python
# Cdigo Python  Ejercicio 1.19
import math

def prob_orden_creciente(k):
    return 1 / math.factorial(k)

# Ejemplo para k = 3 y k = 4
for k in [3, 4]:
    prob = prob_orden_creciente(k)
    print(f"P(orden creciente) para k = {k}: {prob:.4f}")
```

## Resultado final

La probabilidad de que un subconjunto aleatorio de $$k$$ elementos del conjunto $$\{1, \ldots, n\}$$ est ordenado en forma creciente es:

$$
P = \frac{1}{k!}
$$

Este resultado es independiente del valor de $$n$$, siempre que $$k < n$$.

# Ejercicio 1.21  Probabilidades con eventos mutuamente excluyentes

Supn que los eventos $$A$$ y $$B$$ son **mutuamente excluyentes**, con:

$$
P(A) = 0.30, \quad P(B) = 0.60
$$

Calcula:

a) La probabilidad de que ocurra al menos uno de los dos eventos  
b) La probabilidad de que ocurran ambos eventos  
c) La probabilidad de que no ocurra ninguno de los dos eventos  
d) La probabilidad de que ocurra **exactamente uno** de los dos eventos

## Fundamento terico

Dos eventos son **mutuamente excluyentes** si:

$$
P(A \cap B) = 0
$$

Y se cumple que:

$$
P(A \cup B) = P(A) + P(B)
$$

La probabilidad de que **ninguno ocurra** es:

$$
P(\text{ni A ni B}) = 1 - P(A \cup B)
$$

La probabilidad de que **exactamente uno ocurra** es:

$$
P(A \setminus B) + P(B \setminus A) = P(A) + P(B) - 2P(A \cap B)
$$

Pero como $$P(A \cap B) = 0$$, esto se reduce a:

$$
P(\text{exactamente uno}) = P(A) + P(B)
$$

## Resolucin paso a paso

a) Al menos uno de los eventos ocurre:

$$
P(A \cup B) = P(A) + P(B) = 0.30 + 0.60 = 0.90
$$

b) Ambos eventos ocurren:

$$
P(A \cap B) = 0 \quad \text{(porque son mutuamente excluyentes)}
$$

c) Ninguno ocurre:

$$
P(\text{ni A ni B}) = 1 - P(A \cup B) = 1 - 0.90 = 0.10
$$

d) Exactamente uno ocurre:

$$
P(A) + P(B) = 0.30 + 0.60 = 0.90
$$

```python
# Cdigo Python  Ejercicio 1.21
P_A = 0.30
P_B = 0.60
P_A_and_B = 0  # porque son mutuamente excluyentes

# a) Al menos uno
p_union = P_A + P_B

# b) Ambos
p_interseccion = P_A_and_B

# c) Ninguno
p_ninguno = 1 - p_union

# d) Exactamente uno
p_exactamente_uno = P_A + P_B

print(f"a) P(al menos uno) = {p_union:.2f}")
print(f"b) P(ambos) = {p_interseccion:.2f}")
print(f"c) P(ninguno) = {p_ninguno:.2f}")
print(f"d) P(exactamente uno) = {p_exactamente_uno:.2f}")
```

- Una cuarta parte respondi correctamente la **segunda pregunta**:  
  $$P(B) = \frac{1}{4}$$

- Una tercera parte respondi correctamente la **tercera pregunta**:  
  $$P(C) = \frac{1}{3}$$

Adems:

- Una dcima parte respondi correctamente **cada par** de preguntas:  
  $$P(A \cap B) = P(A \cap C) = P(B \cap C) = \frac{1}{10}$$

- Ningn estudiante respondi correctamente **las tres preguntas**:  
  $$P(A \cap B \cap C) = 0$$

Se pide calcular la **probabilidad de que un estudiante no haya respondido correctamente ninguna de las tres preguntas**.

## Fundamento terico

Queremos encontrar:

$$
P(\text{ninguna correcta}) = 1 - P(A \cup B \cup C)
$$

Aplicamos la frmula de inclusin-exclusin para tres eventos:

$$
P(A \cup B \cup C) = P(A) + P(B) + P(C)
- P(A \cap B) - P(A \cap C) - P(B \cap C)
+ P(A \cap B \cap C)
$$

Sustituimos los valores conocidos:

$$
P(A) = \frac{1}{5}, \quad P(B) = \frac{1}{4}, \quad P(C) = \frac{1}{3}
$$

$$
P(A \cap B) = P(A \cap C) = P(B \cap C) = \frac{1}{10}, \quad P(A \cap B \cap C) = 0
$$

## Resolucin paso a paso

Calculamos paso a paso:

$$
P(A \cup B \cup C) = \frac{1}{5} + \frac{1}{4} + \frac{1}{3}
- \frac{1}{10} - \frac{1}{10} - \frac{1}{10} + 0
$$

Convertimos a fracciones comunes:

$$
= \frac{12}{60} + \frac{15}{60} + \frac{20}{60} - \frac{6}{60} - \frac{6}{60} - \frac{6}{60}
= \frac{47}{60} - \frac{18}{60} = \frac{29}{60}

Entonces:

$$
P(\text{ninguna correcta}) = 1 - \frac{29}{60} = \frac{31}{60}

# Ejercicio 1.43  Simulacin: probabilidad de obtener al menos un 8 como suma de dos dados

Simula la probabilidad de que, al lanzar dos dados, **al menos una vez la suma sea igual a 8**.

## Fundamento terico

Cada lanzamiento de dos dados genera un resultado entre 2 y 12.  
La **suma 8** puede obtenerse con las siguientes combinaciones:

$$
(2,6),\ (3,5),\ (4,4),\ (5,3),\ (6,2)
$$

Total de combinaciones posibles al lanzar dos dados:  
$$
6 \times 6 = 36
$$

La probabilidad de obtener un 8 en un solo intento es:

$$
P(\text{suma} = 8) = \frac{5}{36} \approx 0.1389
$$

Para simular **al menos un 8** en mltiples lanzamientos, usamos el **complemento**:

$$
P(\text{al menos un 8 en n lanzamientos}) = 1 - P(\text{ningn 8})^n
$$

En lugar de usar frmula exacta, vamos a estimarla **va simulacin**.

```python
# Cdigo Python  Ejercicio 1.43 (simulacin con Python)
import random

def lanzar_dados():
    return random.randint(1, 6) + random.randint(1, 6)

def simular_al_menos_un_8(n_intentos=10000):
    exitos = 0
    for _ in range(n_intentos):
        if lanzar_dados() == 8:
            exitos += 1
    return exitos / n_intentos

# Simulacin con 10,000 repeticiones
probabilidad_simulada = simular_al_menos_un_8(10000)
print(f"Probabilidad estimada de obtener al menos un 8: {probabilidad_simulada:.4f}")
```

# Ejercicio 2.3  Clculo de probabilidad condicional

Supn que:

$$
P(A) = P(B) = p_1, \quad P(A \cup B) = p_2
$$

Encuentra la probabilidad condicional:

$$
P(A \mid B)
$$

## Fundamento terico

La **probabilidad condicional** se define como:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

Pero no se nos da directamente $$P(A \cap B)$$.  
Vamos a deducirla usando la frmula de la unin:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

Despejamos:

$$
P(A \cap B) = P(A) + P(B) - P(A \cup B)
$$

Como $$P(A) = P(B) = p_1$$:

$$
P(A \cap B) = 2p_1 - p_2
$$

Luego sustituimos en la frmula de la condicional:

$$
P(A \mid B) = \frac{2p_1 - p_2}{p_1}
$$

## Resolucin paso a paso

1. Usamos la frmula de la unin:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

2. Como $$P(A) = P(B) = p_1$$, se tiene:

$$
P(A \cap B) = 2p_1 - p_2
$$

3. Aplicamos la definicin de probabilidad condicional:

$$
P(A \mid B) = \frac{2p_1 - p_2}{p_1}
$$

```python
# Cdigo Python  Ejercicio 2.3
def prob_condicional(p1, p2):
    numerador = 2 * p1 - p2
    denominador = p1
    return numerador / denominador

# Ejemplo con p1 = 0.4, p2 = 0.6
p1 = 0.4
p2 = 0.6
resultado = prob_condicional(p1, p2)
print(f"P(A | B) = {resultado:.4f}")
```

## Resultado final del ejercicio 2.3

La probabilidad condicional es:

$$
P(A \mid B) = \frac{2p_1 - p_2}{p_1}
$$

Ejemplo numrico con $$p_1 = 0.4$$ y $$p_2 = 0.6$$:

$$
P(A \mid B) = \frac{2(0.4) - 0.6}{0.4} = \frac{0.2}{0.4} = 0.5
$$

# Ejercicio 2.9  Probabilidad condicional: flush en pker

En el juego de pker, un **flush** es tener cinco cartas del mismo palo.

Usa **probabilidad condicional** para encontrar la probabilidad de que te repartan un flush en una mano de 5 cartas.

## Fundamento terico

Un mazo estndar de pker tiene **52 cartas**, divididas en **4 palos** (   ) con 13 cartas cada uno.

Total de manos posibles al repartir 5 cartas:

$$
\binom{52}{5}
$$

Para obtener un **flush**, las 5 cartas deben pertenecer al **mismo palo**.

Nmero de maneras de obtener un flush:

- Elegir un palo: 4 formas
- Elegir 5 cartas de ese palo: $$\binom{13}{5}$$

Entonces, nmero total de flushes:

$$
N_{\text{flush}} = 4 \cdot \binom{13}{5}
$$

Probabilidad de un flush:

$$
P(\text{flush}) = \frac{4 \cdot \binom{13}{5}}{\binom{52}{5}}
$$

## Resolucin paso a paso

Calculamos:

- $$\binom{13}{5} = 1287$$
- $$\binom{52}{5} = 2598960$$

Entonces:

$$
P = \frac{4 \cdot 1287}{2598960} = \frac{5148}{2598960} \approx 0.00198
$$

```python
# Cdigo Python  Ejercicio 2.9
import math

# Funcin combinatoria
def combinatoria(n, k):
    return math.comb(n, k)

total_manos = combinatoria(52, 5)
flushes = 4 * combinatoria(13, 5)
p_flush = flushes / total_manos

print(f"Probabilidad de obtener un flush en una mano de 5 cartas: {p_flush:.6f}")
```

$$

**Resultado final:**

# Ejercicio 3.3  Independencia: probabilidad de pudricin de raz

Se sabe que:

- Hay un 70% de probabilidad de que un rbol tenga **pudricin de raz o enfermedad de la corteza**:  
  $$P(R \cup C) = 0.70$$

- La probabilidad de que **no tenga enfermedad de la corteza** es 0.4:  
  $$P(C^c) = 0.40 \Rightarrow P(C) = 1 - 0.40 = 0.60$$

Adems, se indica que tener **pudricin de raz** es **independiente** de tener **enfermedad de la corteza**.

Se pide calcular:

- $$P(R)$$ = probabilidad de que un rbol tenga **pudricin de raz**

## Fundamento terico

Si dos eventos son independientes, entonces:

$$
P(R \cap C) = P(R) \cdot P(C)
$$

Tambin se sabe por frmula de la unin:

$$
P(R \cup C) = P(R) + P(C) - P(R \cap C)
$$

Sustituimos usando la independencia:

$$
P(R \cup C) = P(R) + P(C) - P(R) \cdot P(C)
$$

## Resolucin paso a paso

Sabemos:

- $$P(R \cup C) = 0.70$$  
- $$P(C) = 0.60$$  
- Y que:  
  $$P(R \cup C) = P(R) + 0.60 - P(R) \cdot 0.60$$

Planteamos la ecuacin:

$$
0.70 = P(R) + 0.60 - 0.60 \cdot P(R)
$$

Factorizamos:

$$
0.70 = P(R)(1 - 0.60) + 0.60 = 0.40 \cdot P(R) + 0.60
$$

Restamos 0.60 de ambos lados:

$$
0.10 = 0.40 \cdot P(R) \Rightarrow P(R) = \frac{0.10}{0.40} = 0.25
$$

```python
# Cdigo Python  Ejercicio 3.3
P_union = 0.70
P_C = 0.60

# Resolviendo: 0.70 = P(R) + P(C) - P(R) * P(C)
#             0.70 = R + 0.60 - R*0.60
#             0.10 = R*(1 - 0.60)  R = 0.10 / 0.40

P_R = (P_union - P_C) / (1 - P_C)
print(f"P(R) = {P_R:.4f}")
```

## Resultado final del ejercicio 3.3

La probabilidad de que un rbol tenga **pudricin de raz** es:

$$

# Ejercicio 3.21  Variable aleatoria binomial: infeccin en grupo de personas

En un grupo de 1000 personas:

- Cada persona tiene una probabilidad del 1% de estar infectada por un virus  
  $$P(\text{infectado}) = 0.01$$

- El proceso es **independiente** de una persona a otra

Sea $$X \sim \text{Binomial}(n = 1000,\ p = 0.01)$$

Escribe y resuelve las siguientes probabilidades:

a) $$P(X = 10)$$  
b) $$P(X \geq 16)$$  
c) $$P(12 \leq X \leq 14)$$  
d) $$P(X \geq 1)$$

## Fundamento terico

La distribucin binomial se define como:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

Usaremos la funcin binomial acumulada para resolver los incisos b, c y d:

- $$P(X \geq 16) = 1 - P(X \leq 15)$$  
- $$P(12 \leq X \leq 14) = P(X \leq 14) - P(X \leq 11)$$  
- $$P(X \geq 1) = 1 - P(X = 0)$$

```python
# Cdigo Python  Ejercicio 3.21
from scipy.stats import binom

n = 1000
p = 0.01

# a) P(X = 10)
a = binom.pmf(10, n, p)

# b) P(X >= 16)
b = 1 - binom.cdf(15, n, p)

# c) P(12 <= X <= 14)
c = binom.cdf(14, n, p) - binom.cdf(11, n, p)

# d) P(X >= 1)
d = 1 - binom.pmf(0, n, p)

print(f"a) P(X = 10)         = {a:.4f}")
print(f"b) P(X >= 16)        = {b:.4f}")
print(f"c) P(12 <= X <= 14)  = {c:.4f}")
print(f"d) P(X >= 1)         = {d:.4f}")
```

## Fundamento terico

La funcin de masa de probabilidad de una variable Poisson es:

$$

$$

Para obtener la probabilidad de que $$X$$ sea impar, sumamos sobre los $$k$$ impares:

$$
P(X \text{ impar}) = \sum_{k=1,3,5,\ldots}^{\infty} \frac{e^{-\lambda} \lambda^k}{k!}
$$

Esta suma no tiene una forma cerrada simple, pero puede simularse numricamente para obtener un valor aproximado.

```python
# Cdigo Python  Ejercicio 3.35
import math

def poisson_impar_prob(lambda_, max_k=100):
    prob = 0
    for k in range(1, max_k + 1, 2):  # solo impares
        term = math.exp(-lambda_) * (lambda_**k) / math.factorial(k)
        prob += term
    return prob

# Ejemplo con lambda = 5
lambda_valor = 5
prob_impar = poisson_impar_prob(lambda_valor)
print(f"P(X impar) con  = {lambda_valor}: {prob_impar:.4f}")
```

Como esta suma infinita no puede evaluarse directamente, realizamos una **simulacin Monte Carlo**:

1. Generamos muchas variables $$X_i \sim \text{Poisson}(\lambda)$$
2. Contamos cuntas de ellas son impares
3. Estimamos:

# Ejercicio 4.19  Distribucin conjunta discreta

Se da la siguiente **funcin de masa de probabilidad conjunta**:

$$
\begin{aligned}
P(X = 1, Y = 1) &= \frac{1}{8} \\
P(X = 1, Y = 2) &= \frac{1}{4} \\
P(X = 2, Y = 1) &= \frac{1}{8} \\
P(X = 2, Y = 2) &= \frac{1}{2}
\end{aligned}
$$

Se pide:

a) Encontrar las **distribuciones marginales** de $$X$$ y $$Y$$  
b) Determinar si $$X$$ y $$Y$$ son **independientes**  
c) Calcular $$P(XY \leq 3)$$  
d) Calcular $$P(X + Y > 2)$$

## Fundamento terico

La **distribucin marginal** se obtiene sumando sobre la otra variable:

- Marginal de $$X$$:

$$
P_X(x) = \sum_y P(X = x, Y = y)
$$

- Marginal de $$Y$$:

$$
P_Y(y) = \sum_x P(X = x, Y = y)
$$

La independencia se verifica con:

$$
P(X = x, Y = y) = P_X(x) \cdot P_Y(y) \quad \forall x,y
$$

```python
# Cdigo Python  Ejercicio 4.19
import pandas as pd

# Definimos la distribucin conjunta como tabla
joint = {
    (1, 1): 1/8,
    (1, 2): 1/4,
    (2, 1): 1/8,
    (2, 2): 1/2
}

# a) Distribuciones marginales
marginal_X = {}
marginal_Y = {}

for (x, y), p in joint.items():
    marginal_X[x] = marginal_X.get(x, 0) + p
    marginal_Y[y] = marginal_Y.get(y, 0) + p

# b) Verificar independencia
independiente = True
for (x, y), p in joint.items():
    if abs(p - marginal_X[x] * marginal_Y[y]) > 1e-6:
        independiente = False
        break

# c) P(XY  3)
p_menor_igual_3 = sum(p for (x, y), p in joint.items() if x * y <= 3)

# d) P(X + Y > 2)
p_suma_mayor_2 = sum(p for (x, y), p in joint.items() if x + y > 2)

# Resultados
print("a) Distribucin marginal de X:", marginal_X)
print("   Distribucin marginal de Y:", marginal_Y)
print("b) X y Y son independientes?:", independiente)
print(f"c) P(XY  3) = {p_menor_igual_3:.3f}")
print(f"d) P(X + Y > 2) = {p_suma_mayor_2:.3f}")
```

## Resultado final del ejercicio 4.19

a) Distribuciones marginales:

- $$P_X(1) = \frac{1}{8} + \frac{1}{4} = \frac{3}{8}$$  
- $$P_X(2) = \frac{1}{8} + \frac{1}{2} = \frac{5}{8}$$

- $$P_Y(1) = \frac{1}{8} + \frac{1}{8} = \frac{1}{4}$$  
- $$P_Y(2) = \frac{1}{4} + \frac{1}{2} = \frac{3}{4}$$

b) $$X$$ y $$Y$$ **no son independientes**, ya que:  
$$P(X = 1, Y = 1) = \frac{1}{8} \neq \frac{3}{8} \cdot \frac{1}{4} = \frac{3}{32}$$

c) $$P(XY \leq 3) = \frac{1}{8} + \frac{1}{4} + \frac{1}{8} = \frac{1}{2}$$

d) $$P(X + Y > 2) = \frac{1}{4} + \frac{1}{8} + \frac{1}{2} = \frac{7}{8}$$

# Ejercicio 4.59  Simulacin de la varianza del problema del emparejamiento

Simula la **varianza** del nmero de coincidencias en el **problema del emparejamiento** (matching problem) para un valor grande de $$n$$.

## Fundamento terico

El problema del emparejamiento consiste en lo siguiente:

- Tenemos $$n$$ elementos ordenados (por ejemplo, sobres y cartas)
- Se mezclan aleatoriamente
- Se cuentan los **"matches"**: posiciones donde el elemento coincide con su posicin original

La variable aleatoria $$X_n$$ representa el nmero de coincidencias.  
Para un valor grande de $$n$$, se puede estimar su **esperanza** y **varianza** por simulacin:

- $$E(X_n) = 1$$
- La varianza terica es conocida:  
  $$\text{Var}(X_n) = 1$$ tambin

```python
# Cdigo Python  Ejercicio 4.59 (simulacin del matching problem)
import numpy as np

def simular_emparejamiento(n, repeticiones=100000):
    coincidencias = []
    for _ in range(repeticiones):
        perm = np.random.permutation(n)
        matches = np.sum(perm == np.arange(n))
        coincidencias.append(matches)
    return np.var(coincidencias)

# Simulacin con n grande
n = 1000
var_simulada = simular_emparejamiento(n)
print(f"Varianza simulada para n = {n}: {var_simulada:.4f}")
```

$$

Escribe un script para estimar $$\pi$$ usando el **problema de la aguja de Buffon**.

Pregunta:

Cuntas iteraciones necesitas realizar para estar razonablemente seguro de que la estimacin es buena con al menos **dos cifras significativas** (es decir, que $$\pi \approx 3.14$$)?

## Fundamento terico

El problema de Buffon plantea:

- Se deja caer una aguja de longitud $$L$$ sobre un plano con lneas paralelas separadas por una distancia $$D$$
- Si $$L \leq D$$, la probabilidad de que la aguja cruce una lnea es:

$$
P = \frac{2L}{\pi D}
$$

Por lo tanto, podemos estimar $$\pi$$ con:

$$
\pi \approx \frac{2L \cdot N}{D \cdot C}
$$

donde:

- $$N$$ es el nmero total de lanzamientos
- $$C$$ es el nmero de veces que la aguja cruz una lnea

```python
# Cdigo Python  Ejercicio 6.63 (Buffon)
import numpy as np

def estimar_pi_buffon(n_simulaciones, L=1, D=1):
    x = np.random.uniform(0, D/2, size=n_simulaciones)  # distancia del centro de la aguja a la lnea ms cercana
    theta = np.random.uniform(0, np.pi/2, size=n_simulaciones)  # ngulo con respecto a la lnea

    cruces = np.sum(x <= (L/2) * np.sin(theta))
    if cruces == 0:
        return None  # evitar divisin por cero
    pi_estimado = (2 * L * n_simulaciones) / (D * cruces)
    return pi_estimado

# Ejecutar simulacin con varios tamaos
for n in [1000, 5000, 10000, 50000, 100000]:
    estimacion = estimar_pi_buffon(n)
    print(f"Simulacin con {n:>6} lanzamientos    {estimacion:.5f}")
```

Transformamos a la variable estndar:

$$

Esto nos permite usar la tabla de la normal estndar $$Z \sim \mathcal{N}(0,1)$$

Para cada inciso:

- b) $$P(-9 < X < 1) = P\left(\frac{-9 + 4}{5} < Z < \frac{1 + 4}{5}\right) = P(-1 < Z < 1)$$
- c) $$P(\sqrt{X} > 1)$$

$$

# Ejercicio 7.2  Transformaciones de una variable normal

Sea $$X \sim \mathcal{N}(4, 4)$$, es decir:

- Media: $$\mu = 4$$
- Varianza: $$\sigma^2 = 4 \Rightarrow \sigma = 2$$

Calcula las siguientes probabilidades:

a) $$P(|X| < 2)$$  
b) $$P(e^X < 1)$$  
c) $$P(X^2 > 3)$$

## Fundamento terico

Transformamos los eventos a la forma $$Z \sim \mathcal{N}(0, 1)$$:

$$
Z = \frac{X - \mu}{\sigma}
$$

a) $$P(|X| < 2)$$  
 Resolveremos como:
$$
P(-2 < X < 2) = \Phi\left(\frac{2 - 4}{2}\right) - \Phi\left(\frac{-2 - 4}{2}\right) = \Phi(-1) - \Phi(-3)
$$

b) $$P(e^X < 1)$$  
 Aplicamos logaritmo:
$$
e^X < 1 \Rightarrow X < \ln(1) = 0 \Rightarrow Z = \frac{0 - 4}{2} = -2
$$

c) $$P(X^2 > 3)$$  
 Resolver como:
$$
P(X < -\sqrt{3}) + P(X > \sqrt{3})
$$

```python
# Cdigo Python  Ejercicio 7.2
mu = 4
sigma = 2

from numpy import sqrt, log
from scipy.stats import norm

# a) P(|X| < 2)  P(-2 < X < 2)
z1 = (-2 - mu) / sigma
z2 = (2 - mu) / sigma
p_a = norm.cdf(z2) - norm.cdf(z1)

# b) P(e^X < 1)  X < ln(1) = 0
z_b = (0 - mu) / sigma
p_b = norm.cdf(z_b)

# c) P(X^2 > 3)  P(X < -3 or X > 3)
x = sqrt(3)
z_c1 = (-x - mu) / sigma
z_c2 = (x - mu) / sigma
p_c = norm.cdf(z_c1) + (1 - norm.cdf(z_c2))

print(f"a) P(|X| < 2) = {p_a:.4f}")
print(f"b) P(e^X < 1) = {p_b:.4f}")
print(f"c) P(X^2 > 3) = {p_c:.4f}")
```

# Ejercicio 7.3  Normal: peso al nacer de bebs

Los pesos al nacer de los bebs se distribuyen normalmente con:

- Media: $$\mu = 120$$ onzas  
- Desviacin estndar: $$\sigma = 20$$ onzas

Se pide:

a) Qu peso se considera **bajo peso al nacer**, si est en el **quinto percentil**?  
b) Cul es la probabilidad de que un beb pese **menos de 52 onzas**?  
c) Si se sabe que un beb tiene bajo peso al nacer, cul es la probabilidad de que su peso sea **menor a 52 onzas**?

## Fundamento terico

Sea $$X \sim \mathcal{N}(120,\ 20^2)$$.

Para resolver:

a) Quinto percentil  buscamos el valor $$x$$ tal que:

$$
P(X < x) = 0.05 \Rightarrow x = \mu + z_{0.05} \cdot \sigma
$$

b) Directo con CDF:  
$$
P(X < 52) = \Phi\left(\frac{52 - 120}{20}\right)
$$

c) Probabilidad condicional:

$$
P(X < 52 \mid X < x_{5\%}) = \frac{P(X < 52)}{P(X < x_{5\%})}
$$

```python
# Cdigo Python  Ejercicio 7.3
from scipy.stats import norm

mu = 120
sigma = 20

# a) Peso en el quinto percentil
x_p5 = norm.ppf(0.05, mu, sigma)

# b) P(X < 52)
p_b = norm.cdf(52, mu, sigma)

# c) P(X < 52 | X < p5)
p_c = p_b / 0.05

print(f"a) Peso correspondiente al 5 percentil: {x_p5:.2f} oz")
print(f"b) P(X < 52) = {p_b:.4f}")
print(f"c) P(X < 52 | bajo peso) = {p_c:.4f}")
```

## Resultado final del ejercicio 7.3

a) **Bajo peso al nacer** (percentil 5):

$$
x = 120 + z_{0.05} \cdot 20 \approx 120 - 1.645 \cdot 20 = 87.10 \text{ oz}
$$

b) Probabilidad de que un beb pese menos de 52 onzas:

$$
P(X < 52) \approx 0.0009
$$

c) Probabilidad condicional dado que ya es bajo peso:

$$
P(X < 52 \mid X < 87.10) = \frac{0.0009}{0.05} = 0.0174
$$

# Ejercicio 7.11  Desviacin y media muestral en distribucin normal

Sea $$X_1, X_2, \ldots, X_n$$ una secuencia de variables aleatorias i.i.d. con:

- $$X_i \sim \mathcal{N}(\mu = 5,\ \sigma^2 = 1)$$

Sea:

- $$S_n = X_1 + X_2 + \cdots + X_n$$

Se pide:

a) Calcular $$P(|X_1 - \mu| \leq \sigma)$$  
b) Calcular:  
   $$P\left(\left| \frac{S_n}{n} - \mu \right| \leq \sigma \right)$$  
   para $$n = 9$$

## Fundamento terico

a) Sea $$X_1 \sim \mathcal{N}(5, 1)$$. Queremos:

$$
P(4 \leq X_1 \leq 6) = P\left( \left| X_1 - 5 \right| \leq 1 \right)
$$

Transformamos a $$Z$$:

$$
P(4 \leq X_1 \leq 6) = P\left( \frac{4 - 5}{1} \leq Z \leq \frac{6 - 5}{1} \right) = P(-1 \leq Z \leq 1)
= \Phi(1) - \Phi(-1) \approx 0.6826
$$

---

b) La media muestral:

$$
\bar{X}_n = \frac{S_n}{n}
$$

Distribucin de $$\bar{X}_n$$:

$$
\bar{X}_n \sim \mathcal{N} \left( \mu = 5,\ \frac{1}{n} \right)
$$

Entonces:

$$
P\left( \left| \bar{X}_n - \mu \right| \leq \sigma \right)
= P\left( \left| Z \right| \leq \frac{\sigma}{\sqrt{1/n}} \right)
= P\left( \left| Z \right| \leq \sigma \sqrt{n} \right)
$$

Con $$\sigma = 1, n = 9$$:

$$
\sigma \sqrt{n} = 3
\Rightarrow P(|Z| \leq 3) = \Phi(3) - \Phi(-3) = 0.9987
$$

```python
# Cdigo Python  Ejercicio 7.11
# a) P(|X1 - 5| <= 1)
p_a = norm.cdf(6, 5, 1) - norm.cdf(4, 5, 1)

# b) P(|X - 5| <= 1) con n = 9
sigma_bar = 1 / (9**0.5)
p_b = norm.cdf(5 + 1, 5, sigma_bar) - norm.cdf(5 - 1, 5, sigma_bar)

print(f"a) P(|X1 - 5| <= 1) = {p_a:.4f}")
print(f"b) P(|X - 5| <= 1), n=9 = {p_b:.4f}")
```

# Ejercicio 7.21  Proceso de Poisson: llegadas a clase

Desde las 9:00 a.m., los estudiantes llegan a clase de acuerdo con un proceso de **Poisson con parmetro**:

$$
\lambda = 2 \quad \text{(por minuto)}
$$

La clase comienza a las 9:15 a.m. (15 minutos de observacin). Hay 30 estudiantes.

Responde:

a) Cul es la esperanza y la varianza del nmero de estudiantes que han llegado a las 9:15 a.m.?  
b) Cul es la probabilidad de que hayan llegado al menos 10 estudiantes a las 9:05 a.m.?  
c) Cul es la probabilidad de que el ltimo estudiante llegue tarde?  
d) Supn que **exactamente 6 estudiantes** llegaron tarde.  
   Cul es la probabilidad de que **exactamente 15** hayan llegado antes de las 9:10 a.m.?  
e) Cul es el **tiempo esperado** de llegada del **sptimo estudiante**?

## Fundamento terico

Para un proceso de Poisson con tasa $$\lambda$$ por minuto, se tiene:

- Nmero de llegadas en un intervalo $$t$$ minutos:  
  $$N(t) \sim \text{Poisson}(\lambda t)$$

- El **tiempo de llegada del k-simo evento** sigue una distribucin **Gamma**:

$$
T_k \sim \text{Gamma}(k,\ \lambda)
$$

- La probabilidad de al menos $$k$$ llegadas hasta tiempo $$t$$:

$$
P(N(t) \geq k) = 1 - P(N(t) \leq k - 1)
$$

```python
from scipy.stats import poisson

# Parmetros
lam = 2  # tasa por minuto

# a) Esperanza y varianza hasta 9:15 (15 minutos)
esperanza = lam * 15
varianza = lam * 15

# b) P(al menos 10 llegadas antes de 9:05)  t = 5 min
p_b = 1 - poisson.cdf(9, lam * 5)

# c) P(ltimo estudiante llega tarde) = P(menos de 30 llegan antes de 15 min)
p_c = 1 - poisson.cdf(29, lam * 15)

# d) P(exactamente 15 llegadas antes de 9:10)  t = 10 min
p_d = poisson.pmf(15, lam * 10)

# e) Tiempo esperado del sptimo estudiante
esperado_t7 = 7 / lam

# Mostrar resultados con formato
print(f"a) Esperanza = {esperanza}, Varianza = {varianza}")
print(f"b) P(al menos 10 antes de 9:05) = {p_b:.4f}")
print(f"c) P(ltimo estudiante llega tarde) = {p_c:.4f}")
print(f"d) P(exactamente 15 antes de 9:10) = {p_d:.4f}")
print(f"e) Tiempo esperado del 7 estudiante = {esperado_t7:.2f} minutos")
```

## Resultado final del ejercicio 7.21

a) En 15 minutos:

- Esperanza: $$\mathbb{E}[N(15)] = 2 \cdot 15 = 30$$  
- Varianza: $$\text{Var}(N(15)) = 30$$

b) Probabilidad de que hayan llegado **al menos 10** estudiantes a las 9:05 a.m.:

$$
P(N(5) \geq 10) = 1 - P(N(5) \leq 9) \approx 1 - 0.5830 = 0.4170
$$

c) Probabilidad de que **el ltimo estudiante llegue tarde** (despus de 9:15):

$$
P(\text{menos de 30 llegan antes de 15 min}) = 1 - P(N(15) \geq 30) = 1 - 0.5396 = 0.4604
$$

d) Si 6 estudiantes llegaron tarde, la probabilidad de que **exactamente 15** hayan llegado antes de 9:10 (t = 10):

$$
P(N(10) = 15) \approx 0.0347
$$

e) Tiempo esperado de llegada del **sptimo estudiante**:

$$
\mathbb{E}[T_7] = \frac{7}{2} = 3.5 \text{ minutos}
$$

# Ejercicio 7.45  Simulacin de llegadas telefnicas con Poisson

Supn que las llamadas telefnicas llegan a una **mesa de ayuda (Help Desk)** de acuerdo con un **proceso de Poisson** con:

$$
\lambda = 10 \quad \text{(llamadas por unidad de tiempo)}
$$

Simula la llegada de llamadas telefnicas.

## Fundamento terico

Un proceso de Poisson genera eventos en el tiempo con:

- Intervalos entre eventos que siguen una **exponencial**:
  
$$
T_i \sim \text{Exponencial}(\lambda)
$$

La **hora de llegada acumulada** de cada llamada se obtiene sumando los tiempos entre llegadas.

La simulacin consiste en:

1. Generar tiempos entre llegadas:  
   $$\Delta t_i \sim \text{Exp}(\lambda)$$

2. Calcular las llegadas acumuladas:  
   $$T_i = \sum_{j=1}^{i} \Delta t_j$$

```python
# Cdigo Python  Ejercicio 7.45 (Simulacin de proceso de Poisson)
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
lam = 10  # tasa por unidad de tiempo
n_llamadas = 100  # nmero de llamadas a simular

# Generar tiempos entre llamadas (exponenciales)
tiempos_entre_llamadas = np.random.exponential(1 / lam, size=n_llamadas)

# Calcular tiempos acumulados de llegada
tiempos_llegada = np.cumsum(tiempos_entre_llamadas)

# Mostrar primeras llegadas simuladas
for i in range(10):
    print(f"Llamada {i+1}: llega en el tiempo {tiempos_llegada[i]:.4f}")

# Graficar llegadas acumuladas
plt.step(range(1, n_llamadas + 1), tiempos_llegada, where='post')
plt.xlabel("Nmero de llamada")
plt.ylabel("Tiempo acumulado de llegada")
plt.title("Simulacin de llegadas Poisson ( = 10)")
plt.grid(True)
plt.show()
```

## Resultado final del ejercicio 7.45

Simulamos 100 llamadas que llegan de forma aleatoria con:

- Tasa: $$\lambda = 10$$ llamadas por unidad de tiempo
- Tiempos entre llamadas $$\sim \text{Exponencial}(\lambda = 10)$$
- Tiempos de llegada $$T_i = \sum_{j=1}^i \Delta t_j$$

El grfico muestra el crecimiento acumulado del nmero de llamadas en funcin del tiempo, y las primeras 10 llegadas permiten validar visualmente la simulacin.
